##### Random number Generators
A random number generator is a hardware device (such as RSA tocken) or software algorithm that generates a number that is taken from a limited or unlimited distribution and outputs it. The two main types of random number generators are pseudo random number generators and true random number generators.

__Pseudo Random Number Generators__ : Random number generators are typically software, pseudo random number generators. They rely on algorithms to mimic the selection of a value to approximate true randomness. The outputted values from a pseudo random number are adequate for use in most applications but they should not always be relied on for secure cryptographic implementations.

__True Random Number Generators__ : A true random number generator is cryptographically secure and takes into account physical attributes such as atmospheric or thermal conditions. Such tools may also take into account measurement biases. They may also utilize physical coin flipping and dice rolling processes. A TRNG or HRNG is useful for creating seed tokens.

## Portfolio Optimization Implementation

Portfolio theories guide to select securities (instruments) that will maximize returns and minimize risk

portfolio = portfolio(max{returns}, min{risk})

Usually when we build a portfolio, it is advisable to diversify your assets, or purchase different kinds of assets from different companies. The task of optimizing assets involves balancing the trade-offs between risk and return, where return on a stock is the profits realized after a period of time and risk is the standard deviation in an asset's value. 

Portfolio optimization is an important part of the work done by investment banks and asset management firms. One of the early methods is called mean variance optimization. This approach means that, given a set of portfolios with the same returns, we will select the portfolio with assets that have the least statistical relationship to one another.

__Assets__ are of various kinds. An asset is what we would purchase if we want to invest in a company. These include - 

- Bonds
- Stocks
- Cash
- Real Estate

```python
# Define a get_stock function
def get_stock(ticker):
    data = web.DataReader(f"{ticker}","yahoo",start,end)
    data[f'{ticker}'] = data["Close"]
    data = data[[f'{ticker}']] 
    #print(data.head())
    return data
```

We will define a function to generate a single data frame of stock prices for each stock:

```python
from functools import reduce
def combine_stocks(tickers):
    data_frames = []
    for i in tickers:
        data_frames.append(get_stock(i))        
    df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'], how='outer'), data_frames)
    return df_merged
```

## Risk & Return Analysis
Finding the right methods for portfolio optimization is an important part of the work. One of the early methods is called mean variance optimization. This algorithm may select a portfolio containing assets in tech, retail, healthcare and real estate instead of a single industry like tech.

### Mean Variance Optimization 
This method may not perform very well since it makes many simplifying assumptions, such as returns being normally distributed and the need for an invertible covariance matrix. We will calculate the covariance matrix and returns

```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
mhr = mean_historical_return(portfolio)
cs = CovarianceShrinkage(portfolio).ledoit_wolf()
```

Lets calculate the weights. We will use the max Sharpe statistic

```python
from pypfopt.efficient_frontier import EfficientFrontier

ef_frontier = EfficientFrontier(mhr, cs)
weights = ef_frontier.max_sharpe()

cleaned_weights = ef_frontier.clean_weights()
print(dict(cleaned_weights))
```

```python
stocks = ["PFE", "TSLA", "COST"]
portfolio = combine_stocks(stocks).dropna()
plot(portfolio,'Suggested Portfolio Performance')
```

### Measuring risk of a portfolio
- We don't know what the returns of the portfolio be-uncertainty(or expected return) will be called a random variable
- We will also calcualte the variance which is the spread around the mean to capture the volatility. 
- Varinace of low volatile stock is low and for high risk stocks it is more.

The __Sharpe ratio__ is the ratio between returns and risk. The lower the risk and the higher the returns, the higher the Sharpe ratio. The algorithm looks for the maximum Sharpe ratio, which translates to the portfolio with the highest return and lowest risk. Ultimately, the higher the Sharpe ratio, the better the performance of the portfolio. 

```python
ef_frontier.portfolio_performance(verbose=True)
```

We see that our portfolio performs with an expected annual return of 58 percent. Further, the Sharpe ratio value of 1.48 indicates that the portfolio optimization algorithm performs well with our current data. Of course, this return is inflated and is not likely to hold up in the future.

### Hierarchical Risk Parity
Lets try Hierarchical Risk Parity method which uses single linkage clustering which means the distances between two clusters is defined by a single element pair – those two elements which are closest to each other

We will use returns and run the optimization algorithm to get the weights

### Descriptive Statistics
---------------------------------------

It is time to put some numbers to our analysis. The first layer of analysis would include the following:

* $\mbox{Assets returns} (r_i) = \ln\left( \frac{P_i}{P_{i-1}}\right) $

  $\mbox{Annualized Returns} = \left(\sum_{i=1}^{N} r_i \right) \times 52$ 
  

* $\mbox{Annualized Volatility} = \sqrt {\frac{1}{N}\sum_{i=1}^{N}\left(x_i - \mu \right)} \times \sqrt{52} $

* $\mbox{Sharpe Ratio} = \frac{R_p}{\sigma_p}$

```python
datos_returns = np.log(datos/datos.shift(1))
datos_returns.dropna(inplace=True)

stats = pd.DataFrame()
stats['Annualized Returns(%)'] =datos_returns.mean() * semana *100
stats['Annualized Volatility(%)'] = datos_returns.std() * np.sqrt(semana)*100
stats['Sharpe Ratio'] = stats['Annualized Returns(%)'] /stats['Annualized Volatility(%)']

print(82*'-')
print('Assets Classes Annualized Statistics - full observation period')
stats.style.bar(color=['red','green'], align='zero')
```

### Monte Carlo Simulation:
--------------------------------

No we will simulate a The Markowits efficient frontier as well as inefficiently diversified portfolio by running a random simulation over the weights of each asset in both portfolios


```python
pretsEM = []
pvolsEM = []

prets_noEM = []
pvols_noEM = []

[['AGG','SPY','QQQ','EMB']]
[['AGG','SPY','QQQ','GLD']]

for p in range(2500):
    weights = np.random.random(len(allocation)-1)
    weights /= np.sum(weights)
    pretsEM.append(np.sum(datos_returns[['AGG','SPY','QQQ','EMB']].mean()* weights)* semana)
    pvolsEM.append(np.sqrt(np.dot(weights.T,np.dot(datos_returns[['AGG','SPY','QQQ','EMB']].cov()*semana,
                                                             weights))))
pretsEM = np.array(pretsEM)
pvolsEM = np.array(pvolsEM)

for p in range(2500):
    weights = np.random.random(len(allocation)-1)
    weights /= np.sum(weights)
    prets_noEM.append(np.sum(datos_returns[['AGG','SPY','QQQ','GLD']].mean()* weights)* semana)
    pvols_noEM.append(np.sqrt(np.dot(weights.T,np.dot(datos_returns[['AGG','SPY','QQQ','GLD']].cov()*semana,
                                                             weights))))
prets_noEM = np.array(prets_noEM)
pvols_noEM = np.array(pvols_noEM)

# the charts

fig8 = plt.figure(figsize = (12,16))
plt.subplots_adjust(wspace=.5)
plt.subplot(211)

plt.scatter(pvols_noEM, prets_noEM, c = prets_noEM / pvols_noEM, marker = 'o',cmap='viridis')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label = 'Sharpe Ratio')
plt.title('Monte Carlo Simulation Efficient Frontier with no EM')
plt.subplot(212)

plt.scatter(pvolsEM, pretsEM, c = pretsEM / pvolsEM, marker = 'o',cmap='coolwarm')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label = 'Sharpe Ratio')
plt.title('Monte Carlo Simulation Efficient Frontier with EM')

plt.show();
```
